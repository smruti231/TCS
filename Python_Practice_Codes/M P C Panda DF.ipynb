```python
# Concatenating DataFrame

import pandas as pd
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
          'Age': [27, 24, 22, 32],
          'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
          'Qualification': ['Msc', 'MA', 'MCA', 'Phd']}

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
          'Age': [17, 14, 12, 52],
          'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
          'Qualification': ['B.Tech', 'BA', 'Bcom', 'B.hons']}

df = pd.DataFrame(data1, index=[0,1,2,3])
df1 = pd.DataFrame(data2, index=[4,5,6,7])
print(df, "\n\n", df1)
frames = [df, df1]
res1 = pd.concat(frames)
res1
```

```python
# Merging DataFrame 
import pandas as pd
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
          'Age': [27, 24, 22, 32]}
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification': ['B.Tech', 'B.A', 'Bcom', 'B.hons']}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
print(df, "\n\n", df1)
res = pd.merge(df, df1, on = 'key')
res
```

```python
# creating DataFrame using how in an argument

import pandas as pd
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'key1': ['K0', 'K1', 'K0', 'K1'],
         'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
          'Age': [27, 24, 22, 32]}
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
          'key1': ['K0', 'K0', 'K0', 'K0'],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
        'Qualification': ['B.Tech', 'B.A', 'Bcom', 'B.hons']}
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)
print(df, "\n\n", df1)
res = pd.merge(df, df1, how = 'left', on = ['key', 'key1'])
res
```

```python
# Write a Pandas program to check whether alphabetic values present in a given column of a DataFrame

import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\iasms\\Desktop\\Python Prograamms\\Niyati\\Data Files\\word-prevalence.csv')
print("The Original DataFrame:")
print(df)
print("\nwhether the strings are alphabetic!")
df.dropna(inplace = True)
df['is_word_alphabetic'] = df["Word"].str.isalpha()
print(df)
```

```python
# 3 Write a Pandas program to check whether only numeric values present in a given column of a DataFrame.

import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\iasms\\Desktop\\Python Prograamms\\Niyati\\Data Files\\word-prevalence.csv')
print("The Original DataFrame:")
print(df)
print("\nwhether Count contains numeric value or not?")
df.dropna(inplace = True)
df["Count"]= df["Count"].astype(str)
df['is_count_numeric'] = df["Count"].str.isdigit()
print(df)
```

```python
# 4 Write a Pandas program to check if a specified column starts with a specified string in a DataFrame

import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\iasms\\Desktop\\Python Prograamms\\Niyati\\Data Files\\word-prevalence.csv')
print("The Original DataFrame:")
print(df)
print("\nWhether characters in the string starts from a")
search ="a"
df['is_start_with_a'] = df["Word"].str.startswith(search)
print(df)
```

```python
# 5 Write a Pandas program to swap the cases of a specified character column in a given DataFrame

import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\iasms\\Desktop\\Python Prograamms\\Niyati\\Data Files\\word-prevalence.csv')
print("Original DataFrame:")
print(df)
print("\nAfter applying swapping to the word column")
df.dropna(how ='all', inplace = True)
df["swap_word"] = df["Word"].str.swapcase()
print(df)
```
